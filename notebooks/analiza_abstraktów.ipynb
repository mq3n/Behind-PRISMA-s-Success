```python
import pandas as pd

# Wczytanie plików
human = pd.read_csv("abstract_screening/human_abstracts.csv",sep=";")
gpt = pd.read_csv("abstract_screening/gpt_5_abstract_analysis.csv")

# Normalizacja nazw kolumn dla porównania
human_key = "Title"
gpt_key = "original title"

# Filtrowanie tylko 'yes'

filtered = gpt[gpt["yes/no"].str.strip().str.lower() == "yes"]
gpt = filtered.to_dict(orient="records")

# Usunięcie białych znaków i ujednolicenie wielkości liter (opcjonalnie)
# human[human_key] = human[human_key].str.strip().str.lower()
# gpt[gpt_key] = gpt[gpt_key].str.strip().str.lower()

# Zbiory kluczy
human_titles = set(human[human_key])
#gpt_titles = set(gpt[gpt_key])
gpt_titles = {row[gpt_key] for row in gpt if gpt_key in row and row[gpt_key]}

# Wspólne i różne
common_titles = human_titles & gpt_titles
only_in_human = human_titles - gpt_titles
only_in_gpt = gpt_titles - human_titles

print(f"Liczba wspólnych tytułów: {len(common_titles)}")
print(f"Liczba tylko w human: {len(only_in_human)}")
print(f"Liczba tylko w gpt: {len(only_in_gpt)}")

tp = len(common_titles)
tn = len(only_in_human)
fn = len(only_in_gpt)
fp = 62 - tp - tn - fn
precision = tp/(tp+fp)
recall = tp/(tp+fn)
f1 = 2 * (precision*recall / (precision+recall))

print("FP", fp)
print("Precision", precision)
print("Recall", recall)
print("F1", f1)

# DataFrame'y wynikowe
common_df = human[human[human_key].isin(common_titles)]
only_human_df = human[human[human_key].isin(only_in_human)]
only_gpt_df = gpt[gpt[gpt_key].isin(only_in_gpt)]

# Zapis wyników do plików (opcjonalnie)
# common_df.to_csv("common_rows.csv", index=False)
# only_human_df.to_csv("only_in_human.csv", index=False)
# only_gpt_df.to_csv("only_in_gpt.csv", index=False)

# Podgląd
display(common_df.head())
display(only_human_df.head())
display(only_gpt_df.head())
```
