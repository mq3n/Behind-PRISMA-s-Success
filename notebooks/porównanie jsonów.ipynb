```python
import json

def compare_json_files(file1, file2):
    # Załaduj dane JSON z plików
    with open(file1, 'r', encoding='utf-8') as f1, open(file2, 'r', encoding='utf-8') as f2:
        data1 = json.load(f1)
        data2 = json.load(f2)
    
    # Znajdź wspólne elementy
    common_elements = [item for item in data1 if item in data2]
    
    # Znajdź różnice
    differences_file1 = [item for item in data1 if item not in data2]
    differences_file2 = [item for item in data2 if item not in data1]
    
    return data1, data2, common_elements, differences_file1, differences_file2


# Przykład użycia
file1 = 'title_screening/human_titles_screening.json'
file2 = 'title_screening/o3.json'

file1_elements, file2_elements, common_elements, differences_file1, differences_file2 = compare_json_files(file1, file2)

tp = len(common_elements)
fp = len(differences_file2)
fn = len(differences_file1)
tn = 62 - tp - fp - fn
precision = tp/(tp+fp)
recall = tp/(tp+fn)
#f1 = 2 * (precision*recall / (precision+recall))

print("Pozycji w pliku 1:", len(file1_elements))
print("Pozycji w pliku 2:", len(file2_elements))
print("Wspólne elementy (TP):", tp)
print("Elementy które są w 1 ale nie ma w 2 (FN):", fn)
print("Elementy które są w 2 ale nie ma w 1 (FP):", fp)
print("TN", tn)
print("Precision", precision)
print("Recall", recall)
print("F1", f1)


titles = [item['Title'] for item in differences_file2]
for title in titles:
    print(title)

```
